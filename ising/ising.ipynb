# Ising transfer matrix

The following cell computes the of matrix of the critical Ising model on the square lattice, and its eigenvalues, in order to check correctness of the corresponding C++ code.

```julia
using LinearAlgebra
```

## Size 2

```julia

J = log(1+sqrt(2))/2
eJ = exp(J)
emJ = exp(-J)

R0 = [ eJ 0 emJ 0 ;
        0 eJ 0 emJ;
        emJ 0 eJ 0;
        0 emJ 0 eJ ]

R2 = [ eJ emJ  0   0;
       emJ eJ  0   0;
       0   0  eJ emJ;
       0   0  emJ eJ ]

R1 = Diagonal([eJ, emJ, emJ, eJ])
R3 = R1

T2 = R3*R1*R2*R0

eigs2 = reverse(eigvals(T2))
println(eigs2)
```

```julia
init = [1, 0, 0, -1]
transpose(init)*T2*init
```

## Size 3

```julia
V = [ eJ emJ;
     emJ  eJ ]
H = Diagonal([eJ, emJ, emJ, eJ])
i2 = Matrix(1.0I, 2, 2)
V0 = kron(V, i2, i2)
V1 = kron(i2, V, i2)
V2 = kron(i2, i2, V)

H0 = kron(H, i2)
H1 = kron(i2, H)
H2 = Diagonal([eJ, emJ, eJ, emJ, emJ, eJ, emJ, eJ])

T3 = H2*H1*H0*V2*V1*V0

eigs3 = reverse(eigvals(T3))
println(eigs3)
```
