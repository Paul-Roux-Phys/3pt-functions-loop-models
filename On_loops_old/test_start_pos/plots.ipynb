```julia
using JLD2

C123s = load("results/test_4.jld2", "C123s")
C220s = load("results/test_4.jld2", "C220s")
C000s = load("results/test_4.jld2", "C000s")
C101s = load("results/test_4.jld2", "C101s")
C202s = load("results/test_4.jld2", "C202s")
C303s = load("results/test_4.jld2", "C303s")

c123 = C123s[(0, 1, 0, 1)] + C123s[(0, 1, 1, 0)] + C123s[(1, 0, 0, 1)] + C123s[(1, 0, 1, 0)]
c220 = C220s[(0, 1, 0, 1)] + C220s[(0, 1, 1, 0)] + C220s[(1, 0, 0, 1)] + C220s[(1, 0, 1, 0)]
c000 = C000s[(0, 1, 0, 1)] + C000s[(0, 1, 1, 0)] + C000s[(1, 0, 0, 1)] + C000s[(1, 0, 1, 0)]
c101 = C101s[(0, 1, 0, 1)] + C101s[(0, 1, 1, 0)] + C101s[(1, 0, 0, 1)] + C101s[(1, 0, 1, 0)]
c202 = C202s[(0, 1, 0, 1)] + C202s[(0, 1, 1, 0)] + C202s[(1, 0, 0, 1)] + C202s[(1, 0, 1, 0)]
c303 = C303s[(0, 1, 0, 1)] + C303s[(0, 1, 1, 0)] + C303s[(1, 0, 0, 1)] + C303s[(1, 0, 1, 0)]
ω = c123 / c220 * sqrt(c000 * c202 / c101 / c303) 

println(ω)
```
